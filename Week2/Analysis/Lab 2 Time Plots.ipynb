```python
# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt

# open .xvg file and read data from line 30
with open('COMPLAB/comp-lab-class/Week2/Data/1hz3_T310.run.25000000.energy.xvg') as f:
    data = f.readlines()[29:]
```

```python
# plot data from .xvg file as a function of time (ns) and Potential Energy (kJ/mol) with columns 0 and 1
# title graph "Potential Energy vs. Time"
plt.plot([float(line.split()[0]) for line in data], [float(line.split()[1]) for line in data])
plt.title('Potential Energy vs. Time')
plt.xlabel('Time (ns)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.show()
```

```python
# plot data from .xvg file as a function of time (ns) and Kinetic Energy (kJ/mol) with columns 0 and 2
# title graph "Kinetic Energy vs. Time"
plt.plot([float(line.split()[0]) for line in data], [float(line.split()[2]) for line in data])
plt.title('Kinetic Energy vs. Time')
plt.xlabel('Time (ns)')
plt.ylabel('Kinetic Energy (kJ/mol)')
plt.show()
```

```python
# plot data from .xvg file as a function of time (ns) and Pressure (bar) with columns 0 and 5
# title graph "Pressure vs. Time"
plt.plot([float(line.split()[0]) for line in data], [float(line.split()[5]) for line in data])
plt.title('Pressure vs. Time')
plt.xlabel('Time (ns)')
plt.ylabel('Pressure (bar)')
plt.show()
```

```python
# plot data from .xvg file as a function of time (ns) and Temperature (K) with columns 0 and 4
# title graph "Temperature vs. Time"
plt.plot([float(line.split()[0]) for line in data], [float(line.split()[4]) for line in data])
plt.title('Temperature vs. Time')
plt.xlabel('Time (ns)')
plt.ylabel('Temperature (K)')
plt.show()
```

```python
# plot data from .xvg file as a function of time (ns) and Volume (nm^3) with columns 0 and 6
# title graph "Volume vs. Time"
plt.plot([float(line.split()[0]) for line in data], [float(line.split()[6]) for line in data])
plt.title('Volume vs. Time')
plt.xlabel('Time (ns)')
plt.ylabel('Volume (nm^3)')
plt.show()
```
