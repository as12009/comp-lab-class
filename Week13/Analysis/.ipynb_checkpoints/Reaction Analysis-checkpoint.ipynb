```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1=np.loadtxt("../Outputs/monitor/MONITOR-COLVAR.metadynLog")
```

```python
plt.plot(data1[:,0],data1[:,1])
plt.xlabel('Time (fs)')
plt.ylabel('C-O Distance (A)')
plt.title('Reaction v Time')
plt.savefig('monitor-rxnvtime.png')
plt.show()
```

```python
data2=np.loadtxt("../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog")
```

```python
plt.plot(data2[:,0],data2[:,1])
plt.xlabel('Time (fs)')
plt.ylabel('C-O Distance (A)')
plt.title('Reaction v Time')
plt.show()
plt.savefig('initmonitor-rxnvtime.png')
```

```python
data3=np.loadtxt("../Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog")
plt.plot(data3[:,0],data3[:,1])
plt.xlabel('Time (fs)')
plt.ylabel('C-O Distance (A)')
plt.title('Reaction v Time')
plt.show()
plt.savefig('monitorfreqhills-rxnvtime.png')

```

```python
plt.plot(data1[:,0],data1[:,1], color='red',label="monitor")
plt.plot(data2[:,0],data2[:,1], color='pink', label="metad1_initmonitor")
plt.plot(data3[:,0],data3[:,1], color='purple', label= "metad1_initmonitor_freqhills")
plt.xlabel('Time (fs)')
plt.ylabel('C-O Distance (A)')
plt.title('Reaction v Time')
plt.legend()
plt.show()
plt.savefig('allmonitor-rxnvtime.png')

```
